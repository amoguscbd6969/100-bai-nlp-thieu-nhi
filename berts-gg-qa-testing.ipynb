# berts-gg-qa-testing

Offline Kaggle **submit-only** notebook.

This notebook does not train models. It reads saved test predictions from `/kaggle/working/ggqa_bert_models/artifacts`, builds `submission.csv`, and is ready for Kaggle submission.

```python
import os, json
import numpy as np
import pandas as pd

out_dir = '/kaggle/working/ggqa_bert_models'
art_dir = f'{out_dir}/artifacts'

os.makedirs(art_dir, exist_ok=True)
print('artifact dir:', art_dir)
```

```python
sample = pd.read_csv('/kaggle/input/competitions/google-quest-challenge/sample_submission.csv')

with open(f'{art_dir}/target_cols.json', 'r') as f:
    target_cols = json.load(f)

final_test = np.load(f'{art_dir}/test_final_pp.npy')

assert final_test.shape[0] == len(sample), f'Row mismatch: {final_test.shape[0]} vs {len(sample)}'
assert final_test.shape[1] == len(target_cols), f'Column mismatch: {final_test.shape[1]} vs {len(target_cols)}'

sub = sample.copy()
sub[target_cols] = final_test
sub.to_csv('submission.csv', index=False)
print('saved submission.csv')
sub.head()
```
